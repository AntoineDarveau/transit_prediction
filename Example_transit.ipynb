```python
target_list = ["GJ 436 b",
               "HD 189733 b"]
```

```python
# Define a prediction object
infos = PredictTransit(time_range, target_list, 
                       supp_cols=['pl_orbper','pl_trandur','pl_tranmid','st_j','st_h','ra','dec'])
```

```python
# Get observable events ---- 

# Catch warnings to make sure only one of each appears
with catch_warnings(record=True) as w:
    events = infos.predict()

# Print warnings only once
for w_i in w:
    warn(w_i.message)
```

```python
events.show_in_notebook()
```
