```python
# Get text from a reflink

def get_refname_from_link(link):
    
    out = re.search(">(.*)</a>",link).group(1)
    
    return out.strip()

def get_refname_from_links(links):
    
    return [get_refname_from_link(link) for link in links]
```

```python
def mk_ref_table(table, ref_link=True, ref_col='mpl_reflink'):
    
    # Initiate table with the same structure
    keys = table.keys()
    out = Table(names=keys, dtype=['bytes' for k in keys], masked=True, data=table.mask)
    
    # Get all references
    if ref_link:
        refs = get_refname_from_links(table[ref_col])
    else:
        refs = MaskedColumn(table[ref_col], dtype='bytes')
    
    # Put the ref in each columns
    for key in keys:
        out[key] = refs
        
    out['pl_name'] = table['pl_name']  # Keep the planet name
    out.mask = table.mask  #  Keep mask
    
    return out
```

```python
# Read new database from exoplanet archive 
print("Reading new database...", end="")
new = Table.read("/home/adb/Archive/NasaExoplanetArchive/NExA.ecsv")
new.main_col = "pl_name"
print("Done")

# Read new extended database from exoplanet archive to complete it
print("Reading extended database....", end="")
extended = ExtendedArchive.read("/home/adb/Archive/test/make_big_table/NExA_extended.ecsv")
print("Done")

# Change columns names to fit database
extended.ch_col_names()

# Remove used reference (already in masterfile)
index, = where(extended['mpl_def'])
extended.remove_rows(index)

# Sort to choose the reference accordingly
extended.sort('pl_orbpererr1')

# Group by planet's name
grouped = extended.group_by(['pl_name'])

# Define a reference table in parallel
print("Building new reference table...", end='')
keys = extended.keys()
ref_table = mk_ref_table(new, ref_link=0, ref_col='pl_def_refname')
print("Done")

# Complete new table with extended table
print("Completing database with extended database (may take some time)...", end='')
while len(grouped) > 1:

    grouped = grouped.group_by(['pl_name'])
    index = grouped.groups.indices[:-1]

    # Add to master table
    new = new.complete(grouped[index], 'pl_name', 
                               add_col=False, verbose=False)

    # Add ref to ref_table
    ref = mk_ref_table(grouped[index])
    ref_table = ref_table.complete(ref, 'pl_name', 
                                   add_col=False, verbose=False)

    # Remove from the main table
    grouped.remove_rows(index)
    
print('Done')
```

```python
# Read old table 
old = Table.read("/home/adb/Archive/masterfile.save.ecsv")
old.main_col = "pl_name"

# Read old reference table
old_ref = Table.read("/home/adb/Archive/reference_file.ecsv")
old_ref.main_col = "pl_name"

# Complete old tables with new ones
out = old.complete(new, join_type="outer")
out_ref = old_ref.complete(ref_table, join_type="outer")
```
